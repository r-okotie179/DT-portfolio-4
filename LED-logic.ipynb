```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as ipy
import networkx as nx
```

```python
n = 3
G = nx.grid_2d_graph(n, n)
nx.set_node_attributes(G, False, 'on') # whether the LED turns on
```

```python
nx.draw(G)
```

```python
nodes = G.nodes()
on = []
for node in nodes:
    on.append(nodes[node]['on'])
print(on)
```
